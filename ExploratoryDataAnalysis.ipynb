# **Data Cleaning**

```python
#Find out the no of missing values in each column
missing_values = df.isnull().sum()
```

```python
#sort the missing values in descending order
missing_values = missing_values.sort_values(ascending=False)
#Delete the columns that have no null values
missing_values = missing_values[missing_values!=missing_values.isnull()]
#plot the columns who have missing values
missing_values.plot(kind='barh',colormap='autumn',grid=True,edgecolor='black',linewidth=2,xlabel='Number of missing values',ylabel='Column names',title='Number of missing values for each column')
```

```python
df.isna().sum()
```

```python
# 1. endYear : if it is a film ,I'll fill the endYear column with the StartYear because probably it is the same . if it is a TvShow I'll replace the null value with 'still showing'
# 2. episodes : I'll replace it with "No Data"
# 3. certificates : I'll replace it with "No Certificate"
# 4. num_votes & rating : I'll replace it with "No rate"
# 5. StartYear : I'll replace it with "Unknown"
# 6. plot : I'll replace it with "No Data"
# 7. languages : I'll replace it with "Unknown"
# 8. genres : I'll replace it with "No Genre"
# 9. type : I'll replace it with "No Type"
# 10. runtime : I'll replace it with "Unknown"

df["startYear"] = df["startYear"].fillna("Unknown")
df["episodes"]  = df["episodes"].fillna("No Data")
df["certificate"] = df["certificate"].fillna("No certificate")
df["numVotes"] = df["numVotes"].fillna("No Rate")
df["rating"] = df["rating"].fillna("No Rate")
df["plot"] = df["plot"].fillna("No Data")
df["language"] = df["language"].fillna("Unknown")
df["genres"] = df["genres"].fillna("No Genre")
df["type"] = df["type"].fillna("No Type")
df["runtime"] = df["runtime"].fillna("Unknown")
```

```python
df.type.value_counts()
```

```python
#fill the endyear in the case of a series
df.loc[df["type"].isin(["tvSeries","tvEpisode","tvSpecial","tvMiniSeries"]),'endYear'] = df['endYear'].fillna("Still Showing")

#fill the endyear in case of a film
df.loc[df["type"].isin(["movie","tvMovie","short","video","tvShort","videoGame"]),'endYear'] = df["endYear"].fillna("startYear")
```

```python
df.isnull().sum()
```

```python
df[df.endYear.isnull()]
```

```python
df.loc[(df["type"]== "No Type"), "endYear"] = df["endYear"].fillna("Unknown")
```

```python
df.isnull().sum()
```

Finally, the missing values are effectively handled.

### **Years with more movies and tv shows releases**

```python
#group the dataframe by startyear and count the no of rows for each year
df_startYear = df.groupby("startYear")["title"].count()

#remove the rows with startyear Unkown

df_startYear = df_startYear[df_startYear.index != "Unknown"]

#sort from the first year to last year
df_startYear = df_startYear.sort_index()

years = df_startYear.index.to_list()

#Create the figure anf fix the figure size
plt.figure(figsize=(11.5,6))

#Plot the data
plt.plot(df_startYear[36:], 'r-', marker = '.')

# Customize the plot
plt.title('Number of Movies/TV Shows by Netflix over the past 40 years')
plt.xlabel('Start Year')
plt.ylabel('Count')

#Adjust the x-axis tick spacing
plt.xticks(years[36::2], rotation=45, fontsize=8)

#show the plot
plt.tight_layout()
plt.show()
```

# **Average duration of films and shows on Netflix**

```python
df["runtime"].value_counts()
```

```python
#Let's replace \N value with NaN and make our graph
df.runtime.replace("\\N", np.nan, inplace = True)

#Remove Unknown rows(just 2 rows)
#store movies and tv shows in separate variables
movies = df.loc[(df["runtime"] != "Unknown") & (df["type"].isin(["movie","tvMovie","short","video","tvShort","videoGame"]))]
tv_shows = df.loc[(df["runtime"] != "Unknown") & (df["type"].isin(["tvSeries","tvEpisode","tvSpecial","tvMiniSeries"])) ]

#Convert runtime to float
movies.loc[:,"runtime"] = movies["runtime"].astype(float)
tv_shows.loc[:,"runtime"] = tv_shows["runtime"].astype(float)

#group the dataframe by startyear and find the average runtime
movies_runtime = movies.groupby("startYear")["runtime"].mean()
tv_shows_runtime = tv_shows.groupby("startYear")["runtime"].mean()

#remove the rows with startYear as Unknown
movies_runtime = movies_runtime[movies_runtime.index != "Unknown"]
tv_shows_runtime = tv_shows_runtime[tv_shows_runtime.index != "Unknown"]

#Display just the last 15 years
last_fifteenrow_movies = movies_runtime.iloc[-15:]
last_fifteenrow_shows = tv_shows_runtime.iloc[-15:]

# Plotting the data
plt.plot(last_fifteenrow_movies, 'r--',marker=".", label='Movies')
plt.plot(last_fifteenrow_shows, 'k--',marker=".", label='TV Shows')

# Adding labels and title
plt.xlabel('Start Year')
plt.ylabel('Average Minutes')
plt.title('Average Minutes of Movies and TV Shows in the last 15 years')

# Adding grid lines
plt.grid(True, linestyle='--', alpha=0.5)

# Customizing tick labels
plt.xticks(last_fifteenrow_movies.index.to_list(), rotation=45)

# Adding legend
plt.legend()
plt.tight_layout()

# Display the plot
plt.show()

```

We can note the following:
1. Average minutes of screening movies between 90 and 100 minutes over the years
2. Average minutes of TV show episodes between 40 and 60 minutes per episode

## **Most popular genres of films/shows streamed on Netflix**

```python
df.genres.value_counts().head(20)
```

```python
#replace \N by NaN
df.genres.replace(r"\N",np.nan,inplace=True)
df.genres[df.genres == r'\N'].count()
```

```python
df["genres"] = df["genres"].str.replace(","," & ")
df["genres"]
```

```python
movie_genre = df[df["type"].isin(['movie', 'short', 'tvMovie', 'video', 'videoGame'])]
tv_show_genre = df[df["type"].isin(['tvSeries', 'tvEpisode', 'tvSpecial', 'tvMiniSeries', 'tvShort'])]

#grouping the genre and counting the no of movies and shows in each

movies_genre_count = movie_genre.groupby("genres")["title"].count()
tv_show_genre_count = tv_show_genre.groupby("genres")["title"].count()

#sorting movies_genre_count and tv_show_genre_count in descending order and selecting top 20
movies_genre_count = movies_genre_count.sort_values(ascending = False)[:20]
tv_show_genre_count = tv_show_genre_count.sort_values(ascending = False)[:20]

#get the genre present in both tv shows and movies
common_genre = tv_show_genre_count.index.intersection(movies_genre_count.index)

#filter the genres
movie_genre_common = movies_genre_count.loc[common_genre]
tv_genre_common = tv_show_genre_count.loc[common_genre]

#Create positions for tick labels
x = np.arange(len(common_genre))

#Plotting bar chart
width = 0.35 #width of each bar

fig,ax = plt.subplots(figsize=(10,6))

#Plotting movie genre

ax.barh(x,movie_genre_common, height = width, label = 'Movies', color ="red" )

#Plotting tv shows above movies

ax.barh(x+width , tv_genre_common, height=width, label="TV Shows", color="black" )

#Adding labels and titles
ax.set_xlabel("Count",fontsize=12)
ax.set_ylabel("Genres",fontsize=12)
ax.set_title("Top Genres of Movies and TV Shows")

#Customizing tick labels
ax.set_yticks(x + width/2)
ax.set_yticklabels(common_genre,fontsize=10)

#Adding grid lines
ax.grid(axis='x',linestyle="--",linewidth=0.5)

#Adding legend
ax.legend(fontsize=10)

#Adjusting the layout
plt.tight_layout()

#Display the graph
plt.show()
```


1. The genres differ between movies and TV shows, with each having its own predominant genres.
2. The bar chart primarily showcases Drama, Comedy, and Documentary genres as the most prevalent.

## **Rating and Number of Votes over the years**

```python
#filter and aggregate the data
df.rating = df.rating[df.rating != "No Rate"]

#filter out the rows where the numVotes column is No Rate
df.numVotes = df.numVotes[df.numVotes != "No Rate"]

#filter out the rows with startYear Unknown
df.startYear = df.startYear[df.startYear != "Unknown"]

#Group the filtered data by startYear and find the average rating and sum of numVotes
rate_per_year = df.groupby("startYear").agg({'rating':'mean',"numVotes":"sum"})

#Select data for the last 15 years until 2021
rate_per_year = rate_per_year.iloc[:-1].tail(15)
```

```python
#create the figure object and plot the data
fig,ax1 = plt.subplots(figsize=(11,6))

#plot the rating coulmn as a linechart

ax1.plot(rate_per_year["rating"],label="Ratings",color="black",marker="o",linestyle="-",linewidth=2)

#Set the yaxis label for the line chart
ax1.set_ylabel("Rating")
#Create a second yaxis for the bar chart
ax2=ax1.twinx()

#Plot the numVotes column as a bar chart
ax2.bar(rate_per_year.index,rate_per_year["numVotes"],label="Number of Votes",color="red",alpha=0.7)

# Set the y-axis label for the bar chart
ax2.set_ylabel('Number of Votes')

#set xaxis tick labels
ax1.set_xticks(rate_per_year.index)
ax1.set_xticklabels(rate_per_year.index.astype(int),rotation=45)

# Add a legend to the plot
lines, labels = ax1.get_legend_handles_labels()
bars, bar_labels = ax2.get_legend_handles_labels()
ax1.legend(lines + bars, labels + bar_labels, loc='upper right')

# Add a title
plt.title("The Average Rating with the Number of Votes in the Last 15 Years")

# Add grid lines
plt.grid(True)

# Show the plot
plt.show()
```

The average rating has remained consistently around 6 out of 10 over the past 15 years, regardless of fluctuations in the number of votes.
